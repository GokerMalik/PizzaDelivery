Import libraries that will be needed

Create the Pizza upper class and the sub classes

```python
class Pizza(object):

  def get_description(self):
    return self.description
  
  def get_cost(self):
    return self.cost

class Klasik(Pizza):

  def __init__(self, description, cost):
    self.description = description
    self.cost = cost

class Margarita(Pizza):

  def __init__(self, description, cost):
    self.description = description
    self.cost = cost

class TurkPizzasi(Pizza):

  def __init__(self, description, cost):
    self.description = description
    self.cost = cost

class Dominos(Pizza):

  def __init__(self, description, cost):
    self.description = description
    self.cost = cost
```

Create the Sauce upper class and the sub classes

```python
class Decorator(Pizza):

  def get_cost(self):
    return self.component.get_cost() + \
    Pizza.get_cost(self)

  def get_description(self):
    return self.component.get_description() + \
    ' ' + Pizza.get_description(self)

class Zeytin(Decorator):

  def __init__(self, component, description, cost):
    self.component = component
    self.description = description
    self.cost = cost
  
class Mantar(Decorator):

  def __init__(self, component, description, cost):
    self.component = component
    self.description = description
    self.cost = cost
  
class KeciPeyniri(Decorator):

  def __init__(self, component, description, cost):
    self.component = component
    self.description = description
    self.cost = cost

class Et(Decorator):

  def __init__(self, component, description, cost):
    self.component = component
    self.description = description
    self.cost = cost
  
class Sogan(Decorator):

  def __init__(self, component, description, cost):
    self.component = component
    self.description = description
    self.cost = cost

class Misir(Decorator):

  def __init__(self, component, description, cost):
    self.component = component
    self.description = description
    self.cost = cost
```

Create the Main Loop

```python
validity = 0

PizzaMenusu = open("Menu.txt", "r")
print(PizzaMenusu.read())
PizzaMenusu.close()

while (validity < 6):

  if validity < 1:

    pizzaSecimi = None
    pizza = input(f"\nPizza numarası: ")

    try:
      pizza = int(pizza)
    except:
      pass

    if pizza == 1:
      pizzaSecimi = Klasik("Klasik Pizza", 8)
    elif pizza == 2:
      pizzaSecimi = Margarita("Margarita Pizza", 6)
    elif pizza == 3:
      pizzaSecimi = TurkPizzasi("Türk Pizzası", 10)
    elif pizza == 4:
      pizzaSecimi = Dominos("Dominos Pizza", 7)
      
    if pizzaSecimi != None:
      validity += 1
    else:
      print(f"\nLütfen 1'den 4'e kadar bir numara secin")
    
  if 0 < validity and validity < 2:

    dekor = input (f"\nPizza dekoru:" )
    dekorSecimi = None
    
    try:
      dekor = int(dekor)
    except:
      pass
      
    if dekor == 11:
      dekorSecimi = Zeytin(pizzaSecimi, "zeytinli", 0.5)
    elif dekor == 12:
      dekorSecimi = Mantar(pizzaSecimi, "mantarlı", 0.4)
    elif dekor == 13:
      dekorSecimi = KeciPeyniri(pizzaSecimi, "keçi peynirli", 0.6)
    elif dekor == 14:
      dekorSecimi = Et(pizzaSecimi, "etli", 0.3)
    elif dekor == 15:
      dekorSecimi = Sogan(pizzaSecimi, "soğanlı", 0.2)
    elif dekor == 16:
      dekorSecimi = Misir(pizzaSecimi, "mısırlı", 0.1)

    if dekorSecimi != None:
      validity += 1
    else:
      print(f"\nLütfen 11'den 16'ya kadar bir numara seçin")

  if 1 < validity and validity < 3:

    isim = input(f"\nİsminiz?: ")

    if len(isim) > 0:
      validity +=1
    else:
      print(f"\ndevam etmek için lütfen isim belirtiniz")

  if 2 < validity and validity < 4:

    custId = input(f"\nID Numarası: ")
    custIdInt = None

    try:
      custIdInt = int(custId)
    except:
      pass

    if custIdInt != None:
      validity +=1
    else:
      print(f"\nLütfen ID'nini rakamlardan oluşacak biçimde giriniz")

  if 3 < validity and validity < 5:

    cardNum = input(f"\nKart Numarası: ")
    cardNumInt = None

    try:
      cardNumInt = int(cardNum)
    except:
      pass

    if cardNumInt != None:
      validity +=1
    else:
      print(f"\nLütfen kart numaranızı rakamlardan oluşacak biçimde giriniz")

  if 4 < validity and validity < 6:

    sifre = input(f"\nŞifre: ")
    sifreInt = None

    try:
      sifreInt = int(sifre)
    except:
      pass

    if cardNumInt != None:
      validity +=1
    else:
      print(f"\nLütfen şifrenizi rakamlardan oluşacak biçimde giriniz")
```

Save the information

```python
current_time = datetime.datetime.now()

entryList = [isim, custIdInt, cardNumInt, dekorSecimi.get_description(), dekorSecimi.get_cost(), current_time, sifreInt]


with open('Orders_Database.csv', 'a') as DATABASE:
 
    writerObj = csv.writer(DATABASE)
    writerObj.writerow(entryList)
 
    # Close the file object
    DATABASE.close()
```
